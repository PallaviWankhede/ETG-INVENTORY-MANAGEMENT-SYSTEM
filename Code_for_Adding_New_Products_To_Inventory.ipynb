# Code To Add New Items in the INVENTORY :

```python
fd = open("record.json",'r')
r = fd.read()
fd.close()

import json
record = json.loads(r)
```

```python
total_records = len(record) + 1
print("Next New Product_No in our Inventory is : ", total_records , "\nNew Products Product_Id is : " , total_records + 100)
prod_no = str(input("Enter the Product No :"))
prod_id = int(input("Enter the Product Id :"))
name = str(input("Enter Product_Name :"))
category = str(input("Enter the Product Category :"))
expiry = str(input("Enter the Expiry Date :"))
price = int(input("Enter Price per unit :"))
quantity = int(input("Enter the Total no of Units :"))
total = price*quantity
print("-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------")
print(" Total Amount is :",total)
print("-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------")


key = record.keys()
length = len(key)
lengths = length + 1


if int(prod_no) in range(1,lengths):
  print("-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------")
  record[prod_no] = {"Product_Id": prod_id, "Product_Name": name, "Product_Category": category, "Total_Units": record[prod_no]["Total_Units"] + quantity   , "Price_per_unit": price, "Total_Amount": record[prod_no]["Total_Amount"] + total, "Expiry Date": expiry}
  print("\n Product with Product_No " + prod_no + " updated in the Inventory Successfully!!!")
  print("------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------")
  print(" Details of the Updated Product is as Follows : \n " , record[prod_no])
  print("************************************************************************************************************************************************************************************************")
  print("                *****************  THANK YOU FOR UPDATING THE PRODUCTS IN INVENTORY  *****************                           ")
  print("************************************************************************************************************************************************************************************************")

else:
  record[prod_no] = {"Product_Id" : prod_id, "Product_Name" : name, "Product_Category" : category, "Total_Units" : quantity, "Price_per_unit" : price, "Total_Amount" : total, "Expiry Date" : expiry}
  print("------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------")
  print("\n New Product with Product_No " + prod_no + " added to the Inventory Successfully!!!")
  print("------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------")
  print(" Details of the New added Product is as Follows : "  , record[prod_no])
  print("\n\n ************************************************************************************************************************************************************************************************")
  print("                *****************  THANK YOU FOR ADDING NEW PRODUCTS IN INVENTORY  *****************                           ")
  print("************************************************************************************************************************************************************************************************")

 

js = json.dumps(record)

fd = open("record.json",'w')
fd.write(js)
fd.close()
```
