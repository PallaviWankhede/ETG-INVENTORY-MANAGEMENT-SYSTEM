# CODE TO PURCHASE AN ITEM FROM INVENTORY AND CODE TO CREATE "SALES.JSON" FILE:

```python
import json

fd = open("record.json",'r')
r = fd.read()
fd.close()

records = json.loads(r)
```

```python
import time
times = time.ctime()
customer_name = str(input("Enter your Name : "))
key = records.keys()
length = len(key)
lengths = length + 1
size_of_dct = len(records)

for i in range (1,20):
  print("--------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------")
  print(" Enter 1 - To see all the Items in Inventory. \n Enter 2 - To Purchase an Item from Inventory.  \n Enter 3 - To Exit.")
  print("--------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------")
  option = int(input("Enter your option : "))
  print("--------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------")
  
  if (option == 1):
    print("\n Whole Inventory is as follows :  \n\n", records)
    print("--------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------")
  
  if (option == 2):
    print("                   ****************************** To Purchase the Item from Inventory ******************************             \n")
    no_of_prod = str(input("Enter the Product_No of the Product to Purchase : "))
    print("---------------------------------------------------------------------------------------------------------------------------------------------------")
    no_pro = int(no_of_prod)
    if (no_pro <= size_of_dct):
      print("Details of the Product with Product_No " + no_of_prod + " is as follows :")
      print("---------------------------------------------------------------------------------------------------------------------------------------------------")
      print("Product_Id       : " ,records[no_of_prod]['Product_Id'])
      print("Product_Name     : " ,records[no_of_prod]['Product_Name'])
      print("Product_Category : " ,records[no_of_prod]['Product_Category'])
      print("Expiry_Date      : " ,records[no_of_prod]['Expiry Date'])
      print("Price_per_unit   : " ,records[no_of_prod]['Price_per_unit'])
      print("Total_Units      : " ,records[no_of_prod]['Total_Units'])
      print("Total_Amount     : " ,records[no_of_prod]['Total_Amount'])
      print("---------------------------------------------------------------------------------------------------------------------------------------------------")
      
      whole_units = records[no_of_prod]['Total_Units']
      whole = str(whole_units)
      no_of_quantity = int(input("\nEnter the No. of Items to Purchase : "))
      if (no_of_quantity <= whole_units):
        bill = records[no_of_prod]['Price_per_unit'] * no_of_quantity
        print("Billing Amount is : " , bill )
        print("---------------------------------------------------------------------------------------------------------------------------------------------------")
    
      if (no_of_quantity > whole_units):
        print("Total No. of Items are " + whole + ".Please enter no. of units under " + whole +".")
        break
        
      print(" Enter 1 - To Confirm Order and Print Bill. ")
      print(" Enter 2 - To Cancel Order.")
      choice = int(input("Enter Your Choice : "))
      if (choice == 1):
        print("\n         ******************************************   Your Bill is   **********************************************                       \n\n")
        print("Name of Customer     : " , customer_name)
        print("Product_Id           : ", records[no_of_prod]['Product_Id'])
        print("Product_Name         : ", records[no_of_prod]['Product_Name'])
        print("Product_Category     : ", records[no_of_prod]['Product_Category'])
        print("Expiry_Date          : ", records[no_of_prod]['Expiry Date'])
        print("Price_per_unit       : ", records[no_of_prod]['Price_per_unit'])
        print("Units You Purchased  : ", no_of_quantity)
        print("--------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------")
        print("\n\nYour Billing Amount is : ", records[no_of_prod]['Total_Amount'])
        print("Time : " , times )
        print("\n                   ********************* Thank You !!! ***********************                                       ")
        print("--------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------")
        print("\n ***************************   !!! Do Visit Our Shop Soon !!!   ******************************** ")
        print("--------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------")
        # CODE TO CREATE SALES.JSON FILE :
        #By the below code you can update all the names of the customers with the name and detail of the product they purchased with time and billing amount. 
        import json
        ab = open("sales.json",'r')
        rs = ab.read()
        ab.close()
        sale = json.loads(rs)
        no = len(sale)
        noos = str(no) 
        nos = no + 1
        new = sale[noos]["Customer Id"] + 1
        sale[nos] = {"Customer Id": new ,"Customer Name":customer_name,"Product Id":records[no_of_prod]['Product_Id'],"Product Name":records[no_of_prod]['Product_Name'],"Product Category":records[no_of_prod]['Product_Category'],"Price Per Unit":records[no_of_prod]['Price_per_unit'],"No of Quantity":no_of_quantity,"Billing Amount":bill,"Time and Date":times}
        jsn =json.dumps(sale)
        ab = open("sales.json",'w')
        ab.write(jsn)
        ab.close()

      if (choice == 2):
        print("\nYour order is Cancelled Successfully!!!")
      if (choice >= 3):
        print("--------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------")
        print("No Item Exists")
        print("--------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------")
    
    if (no_pro > size_of_dct):
      print("No Item Exists")
  
  if (option == 3):
    print("\n                   ********************* Thank You !!! ***********************                                       ")
    print("--------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------")
    print("\n ***************************   !!! Do Visit Our Shop Soon !!!   ******************************** ")
    break

  if (option >= 4):
    print("--------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------")
    print("\n  Please Enter Valid Options ")
    print("--------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------")
    

records[no_of_prod]['Total_Units'] = records[no_of_prod]['Total_Units'] - no_of_quantity
records[no_of_prod]['Total_Amount'] = records[no_of_prod]['Total_Amount'] - bill
```
